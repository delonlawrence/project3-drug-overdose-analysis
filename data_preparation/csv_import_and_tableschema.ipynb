```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
# Establish path to csv file
path = "overdose_datasets/drug_deaths_CT.csv"
# Read the data from the csv file to a pandas dataframe
df = pd.read_csv(path)
```

```python
# Establish the local database url that we want to upload the dataframe to

database_url = "postgresql://postgres:postgres@localhost:5432/ct_overdose_deaths"
```

```python
# Establish connection to local database url
engine = create_engine(database_url)
```

```python
# Establish the table name for the new table that will be created with the csv data
table_name = "ct_overdose_deaths"
# upload pandas dataframe to the address listed above to create table with columns in pgAdmin
df.to_sql(table_name, engine, index=False, if_exists="replace")
```

```python
# Visualize data in notebook
conn = engine.connect()
```

```python
# Visualize data in notebook
data = pd.read_sql("SELECT * FROM ct_overdose_deaths", conn)
```
